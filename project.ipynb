```python
import pandas as pd
import csv
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv ('train.csv')
test = pd.read_csv ('test.csv')
display(df)
display(test)
```

```python
df1=df[df["label"]==1]
df2=df[df["label"]==2]
df3=df[df["label"]==3]
df4=df[df["label"]==4]
df5=df[df["label"]==5]
d1={}
d2={}
d3={}
d4={}
d5={}
v=[] 
```

```python
def wrdNum(d, data):
    for itm in data["query"]:
        itm=itm.replace(".","")
        itm=itm.replace("؟","")
        itm=itm.replace("!","")
        itm=itm.replace("،","")
        sentence=itm.split()
        for wrd in sentence:
            if wrd not in v:
                v.append(wrd)
            if wrd in d:
                d[wrd] = d[wrd] + 1
            else:
                d[wrd] = 1
def cntr(d):
    cntr=0
    for key in list(d.keys()):
        cntr = cntr+d[key]
    return cntr
```

```python
wrdNum(d1,df1)
wrdNum(d2,df2)
wrdNum(d3,df3)
wrdNum(d4,df4)
wrdNum(d5,df5)
# print(d1)
```

```python
cnt_d1=cntr(d1)
cnt_d2=cntr(d2)
cnt_d3=cntr(d3)
cnt_d4=cntr(d4)
cnt_d5=cntr(d5)
# print(cnt_d5)
```

```python
p1=len(df1.index)/len(df.index)
p2=len(df2.index)/len(df.index)
p3=len(df3.index)/len(df.index)
p4=len(df4.index)/len(df.index)
p5=len(df5.index)/len(df.index)
lenV=len(v)
print(lenV)
```

```python
def findProb(v,wrd_label,class_cnt,lenV):
    prob={}
    for i in v:
        if i in wrd_label:
            x=((wrd_label[i]+1)/(lenV+class_cnt))
        else:
            x=((1)/(lenV+class_cnt))
        prob[i]=x
    return prob
 
```

```python
prob_per_wrd_d1=findProb(v,d1,cnt_d1,lenV)
prob_per_wrd_d2=findProb(v,d2,cnt_d2,lenV)
prob_per_wrd_d3=findProb(v,d3,cnt_d3,lenV)
prob_per_wrd_d4=findProb(v,d4,cnt_d4,lenV)
prob_per_wrd_d5=findProb(v,d5,cnt_d5,lenV)
prob_per_wrd_d4
```

```python
def classChooseProb(p,text,prob_per_wrd,v,class_cnt):
    label=p
    for i in text:
        if i in prob_per_wrd:
            label*=prob_per_wrd[i]
        else:
            x=((1)/(v+class_cnt))
            label=label*x
    return label
```

```python
def maxProb(p1,p2,p3,p4,p5,text,prob_per_wrd_d1,prob_per_wrd_d2,prob_per_wrd_d3,prob_per_wrd_d4,prob_per_wrd_d5,V,cnt_d1,cnt_d2,cnt_d3,cnt_d4,cnt_d5):
    
    lb1=classChooseProb(p1,text,prob_per_wrd_d1,V,cnt_d1)
    lb2=classChooseProb(p2,text,prob_per_wrd_d2,V,cnt_d2)
    lb3=classChooseProb(p3,text,prob_per_wrd_d3,V,cnt_d3)
    lb4=classChooseProb(p4,text,prob_per_wrd_d4,V,cnt_d4)
    lb5=classChooseProb(p5,text,prob_per_wrd_d5,V,cnt_d5)
    maxp=lb1
    label=1
    if maxp<lb2:
        maxp=lb2
        label=2
    if maxp<lb3:
        maxp=lb3
        label=3
    if maxp<lb4:
        maxp=lb4
        label=4
    if maxp<lb5:
        maxp=lb5
        label=5
    return label
```

```python
res=[]

def finalAns(data):
    
    for itm in data["query"]:
        itm=itm.replace(".","")
        itm=itm.replace("؟","")
        itm=itm.replace("!","")
        itm=itm.replace("،","")
        sentence=itm.split()
        label=maxProb(p1,p2,p3,p4,p5,sentence,prob_per_wrd_d1,prob_per_wrd_d2,prob_per_wrd_d3,prob_per_wrd_d4,prob_per_wrd_d5,lenV,cnt_d1,cnt_d2,cnt_d3,cnt_d4,cnt_d5)
        res.append(label)

finalAns(test)
display(test)
with open('result.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)
    id=0
    header=['id','label']
    data=[]
    writer.writerow(header)
    for i in res:
        x=[id,i]
        data.append(x)
        id+=1
    writer.writerows(data)
```
